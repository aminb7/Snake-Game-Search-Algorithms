# SEARCH <br>
    
## Introduction
In this project we have a problem very similar to the snake game which we are to solve it using different search<br>
strategies.<br>
The goal is for all seeds to be eaten by the snake. An we try to find optimal path to goal<br>
We use BFS, IDS, A star, Weighted Astart to solve this problem and then we have a comparison between them.<br>
A star implemented using 2 differnet heuristics. And we try weighted A star with 2 values for alpha.

=====================================================================================================================

## Modeling
### State
We define state as status of snake and status of seeds. More precisely, the position of snake body and position <br>
of seeds. We store this information in 'State' class. According to this definition the goal state is a state <br>
with no seeds. An it's obvious that there are lots of different goal states to reach. <br>
### Initial state
The initial state of problem is determined in input. And it includes position of snake head(Initialy <br>
length of snake is 1) and positions of seeds.<br>
### Actions
Basically in every state there is 4 possible actions and that is movement to 4 directions (Up, Right, Down <br>
and Left). But in the cases which the snake bites itself, its moves decreases and be more limited.
### Transition between states based on actions
When the snake moves to a direction one tile will be added to its body as its new head. And if it does not eat<br>
any seeds, one last tile of its body will be removed. Otherwise it remains(this happen in the next action of which<br>
the snake eat the seed). Eating 1 point seeds removes them from the map. And eating 2 point seeds couse them to<br>
become a 1 point seed.

=====================================================================================================================

## Algorithms

### BFS
In this strategy shallowest nodes will be expanded. And frontier is a FIFO queue.<br>
It has completeness. Time complexity is O(b^d) and space complexity is O(b^d).<br>
It has optimality.<br>
BFS takes lots of space.

### IDS
This strategy combines the best of breadth-first and depth-first search strategies.<br>
It use DFS as a subroutine. <br>
It has completeness. Time complexity is O(b^d) and space complexity is O(bd).<br>
It has optimality.<br>
Takes less space than BFS and that is its advantage. 

### A*
This strategy assume cost to reach current state and estimated cost to<br>
reach the goal. So it avoids expanding nodes which are leading us to <br>
expensive paths. With good heuristics search will be really fast.<br>
It has completeness. Time complexity and space complexity are exponensial.<br>
Optimality:<br>
- In tree search A* is optimal if heuristic is admissible
- In graph search A* is optimal if heuristic is consistent.

#### heuristics:
We implement A* with 2 different heuristics: <br>
- ##### 1st heuristic: number of remaining seeds
This heuristic is selected by relaxing problem the way<br>
that we can jump to each of seeds and eat them one by one.<br>
###### Admissibility: 
Because of game ruls we the snake could not <br>
eat more than one seeds in one move. Hence if we have n seeds<br>
in map, at least n moves is essential. So the heuristic of each<br>
state is less or equal than actual cost to reach the goal.<br>
So this heuristic is admissible.<br>
###### Consistency:
In each move snake eat a seed or not. So number of seeds<br>
in each move may decrease by 1 or remains at its value.<br>
Hence difference of heuristics of 2 continueous states <br>
is 1 or 0. We also know that each move has real cost of 1<br>
which is equal or more than difference of heuristics of <br>
2 continueous states. So this heuristic is consistent.

- ##### 2st heuristic: Optimal distance(manhattan) between head of snake and the furthest seed
This heuristic is selected by relaxing problem the way<br>
that through the way of reaching the furthest seed<br>
snake can eat all another seeds and finally when it<br>
reach the furthest one the game is finished.
###### Admissibility: 
It is obvious that it is nessesaru to eat the furthest seed in the map. <br>
So real cost of finshing the game is equal or more than optimal manhattan <br>
distance to reach the furthest seed in the map. Hence the heuristic is admissible.<br>
###### Consistency:
In each move manhattan distance of all paths can increase by 1, <br>
decrease by 1 or remains at its value. furthest path is same. <br>
In fact when snake moves from a state to another, manhattan<br>
distance to furthest seed can increase by 1,decrease by 1 or <br>
remains at its value.  We also know that each move has real cost<br>
of 1 which is equal or more than difference of heuristics of <br>
2 continueous states. So this heuristic is consistent.

#### Comparison between heuristics.
We know when heuristics get closer to the real cost, fewer<br>
nodes will expand but cost of calculation of heuristic will<br>
increase. So this is a trade-off between quality of estimate<br>
and work per node.<br>
We spend O(1) to compute heuristic 1. Because computing number<br>
of seeds which is a set in python requiers O(1)<br>
Computing heuristic 2 is O(number of seeds).<br>
Because find furthest seed to reach requires<br>
O(number of seeds).<br>
Heuristic 2 is closer than heuristic 1 to real cost<br>
Heuristic 2 also needs more time to compute. <br>
Actually in 2nd heuristic we expand fewer nodes<br>
But expansion of each node takes more time. <br>

In test 1 they are almost same. <br>
In test 2 2nd heuristic operates better (1/5 time of 1st heuristic) <br>
In test 3 1st heuristic operates better (1/2 time of 2nd heuristic) <br>

### Weighted A* (heuristic: number of remaining seeds)
The idea of weighted A* is speed up search at the exponse of <br>
optimality. Multipling heuristics may cause it to become more<br>
than real cost and discard its admissibility. But fewer nodes<br>
tend to expand.
In this project we implement weughted A* with 1.8 and 4 as the<br>
alpha.<br>
Both 1.8 and 4 specially 4 as our alpha speed of the search <br>
considerably as we can see that in the results. Optimality <br>
is preserved in alpha of 1.8. But alpha of 4 makes A* to give<br>
us a little bit higher answer than optimal answer as we can <br>
in the results. We can also see much fewer nodes expanded and<br>
seen states are a lot fewer in comparison to normal A*.

| --- | --- | --- | --- | --- | --- |

```python
import copy
import time

class Problem:
    def get_input2(self, file_path):
        file = open(file_path)
        map_height, map_width = map(int,file.readline().split(','))
        
        snake_head = tuple(map(int,file.readline().split(',')))
        snake_body = [snake_head]
        
        seed_number = int(file.readline())
        
        seeds = set()
        for i in range(seed_number):
            seeds.add(tuple(map(int,file.readline().split(','))))
        
        self.initial_state = State(snake_body, seeds, map_width, map_height, False)
    
    def get_input1(self):
        map_width, map_height = map(int,input().split(','))
        
        snake_head = tuple(map(int,input().split(',')))
        snake_body = [snake_head]
        
        seed_number = int(input())
        
        seeds = set()
        for i in range(seed_number):
            seeds.add(tuple(map(int,input().split(','))))
        
        self.initial_state = State(snake_body, seeds, map_width, map_height, False)
    
    def goal_test(self, state):
        return len(state.seeds) == 0

    
class State:
    def __init__(self, snake_body, seeds, map_width, map_height, grow):
        self.snake_body = snake_body
        self.seeds = seeds
        self.map_width = map_width
        self.map_height = map_height
        self.grow = grow
    
    def __eq__(self, other):
        return self.snake_body == other.snake_body
    
    def __hash__(self):
        return hash((str(self.snake_body)))
    
    def self_biting(self, action):
        snake_length = len(self.snake_body)
        
        if action[1] == 0:
            if self.snake_body[0][0] + action[0] < 0:
                new_head = (self.map_height - 1, self.snake_body[0][1])

            elif self.snake_body[0][0] + action[0] >= self.map_height:
                new_head = (0, self.snake_body[0][1])

            else:
                new_head = (self.snake_body[0][0] + action[0], self.snake_body[0][1])

        else:
            if self.snake_body[0][1] + action[1] < 0:
                new_head = (self.snake_body[0][0], self.map_width - 1)

            elif self.snake_body[0][1] + action[1] >= self.map_width:
                new_head = (self.snake_body[0][0], 0)
                
            else:
                new_head = (self.snake_body[0][0], self.snake_body[0][1] + action[1])
        
        return ((new_head in self.snake_body) and new_head != self.snake_body[snake_length - 1]) \
                or ((new_head in self.snake_body) and new_head == self.snake_body[snake_length - 1] \
                and snake_length == 2)
        
    def evaluate_actions(self):
        actions = []
        if not self.self_biting((-1, 0)):
            actions.append((-1, 0))
        
        if not self.self_biting((0, +1)):
            actions.append((0, +1))
        
        if not self.self_biting((+1, 0)):
            actions.append((+1, 0))
        
        if not self.self_biting((0, -1)):
            actions.append((0, -1))
        return actions
    
    def get_next_state(self, action):
        snake_body = copy.deepcopy(self.snake_body)
        seeds = copy.deepcopy(self.seeds)

        if action[1] == 0:
            if snake_body[0][0] + action[0] < 0:
                new_head = (self.map_height - 1, self.snake_body[0][1])
                    
            elif snake_body[0][0] + action[0] >= self.map_height:
                new_head = (0, self.snake_body[0][1])
                            
            else:
                new_head = (self.snake_body[0][0] + action[0], self.snake_body[0][1])
                
        else:
            if snake_body[0][1] + action[1] < 0:
                new_head = (self.snake_body[0][0], self.map_width - 1)

            elif snake_body[0][1] + action[1] >= self.map_width:
                new_head = (self.snake_body[0][0], 0)

            else:
                new_head = (self.snake_body[0][0], self.snake_body[0][1] + action[1])
        
        grow = False
        snake_body.insert(0, new_head)
        if (new_head[0], new_head[1], 1) in seeds:
            seeds.remove((new_head[0], new_head[1], 1))
            grow = True
        elif (new_head[0], new_head[1], 2) in seeds:
            seeds.remove((new_head[0], new_head[1], 2))
            seeds.add((new_head[0], new_head[1], 1))
            grow = True
        else:
            snake_body.pop()

        return State(snake_body, seeds, self.map_width, self.map_height, grow)

    
class Node:
    def __init__(self, state, parent, path_cost):
        self.state = state
        self.parent = parent
        self.last_action = None
        self.path_cost = path_cost

    def set_last_action(self, last_action):
        self.last_action = last_action

def print_frontier(frontier):
    for node in frontier:
        print("Snake body: ", node.state.snake_body)
        print("Seeds: ", len(node.state.seeds))
        print("Actions: ", node.state.evaluate_actions())
        print("Cost: ", node.path_cost)
        print("___________")
        
def print_path(node):
    
    parent = node
    path = ""
    while parent:
        if parent.last_action == (-1, 0):
            path += "U-"
        elif parent.last_action == (0, 1):
            path += "R-"
        elif parent.last_action == (1, 0):
            path += "D-"
        elif parent.last_action == (0, -1):
            path += "L-"
        parent = parent.parent
    
    print(path[:-1][::-1])
        
def BFS(problem):
    root = Node(problem.initial_state, None, 0)
    
    if problem.goal_test(root.state):
        print("Path length: ", root.path_cost)
        print("Path: ", end="")
        print_path(root)
        return True
    
    frontier = [root]
    explored = set()
    
    while frontier:
        node = frontier.pop(0)
        
        if node.state not in explored:
            explored.add(node.state)
            
            for action in node.state.evaluate_actions():
                child = Node(node.state.get_next_state(action), node, node.path_cost + 1)
                child.set_last_action(action)
                if problem.goal_test(child.state):
                    print("Path length: ", child.path_cost)
                    print("Path: ", end="")
                    print_path(child)
                    return True
                frontier.append(child)
    return False

    
problem = Problem()
problem.get_input2("test3.txt")

time1 = time.time()
BFS(problem)
time2 = time.time()

print("Time: ", (time2 - time1)*1000, " ms")
```

```python
import copy
import time

class Problem:
    def get_input2(self, file_path):
        file = open(file_path)
        map_height, map_width = map(int,file.readline().split(','))
        
        snake_head = tuple(map(int,file.readline().split(',')))
        snake_body = [snake_head]
        
        seed_number = int(file.readline())
        
        seeds = set()
        for i in range(seed_number):
            seeds.add(tuple(map(int,file.readline().split(','))))
        
        self.initial_state = State(snake_body, seeds, map_width, map_height)
    
    def get_input1(self):
        map_height, map_width = map(int,input().split(','))
        
        snake_head = tuple(map(int,input().split(',')))
        snake_body = [snake_head]
        
        seed_number = int(input())
        
        seeds = set()
        for i in range(seed_number):
            seeds.add(tuple(map(int,input().split(','))))
        
        self.initial_state = State(snake_body, seeds, map_width, map_height)
    
    def goal_test(self, state):
        return len(state.seeds) == 0

    
class State:
    def __init__(self, snake_body, seeds, map_width, map_height):
        self.snake_body = snake_body
        self.seeds = seeds
        self.map_width = map_width
        self.map_height = map_height
    
    def __eq__(self, other):
        return self.snake_body == other.snake_body and self.seeds == other.seeds
    
    def __hash__(self):
        return hash((str(self.snake_body), str(self.seeds)))
    
    def self_biting(self, action):
        snake_length = len(self.snake_body)
        
        if action[1] == 0:
            if self.snake_body[0][0] + action[0] < 0:
                new_head = (self.map_height - 1, self.snake_body[0][1])

            elif self.snake_body[0][0] + action[0] >= self.map_height:
                new_head = (0, self.snake_body[0][1])

            else:
                new_head = (self.snake_body[0][0] + action[0], self.snake_body[0][1])

        else:
            if self.snake_body[0][1] + action[1] < 0:
                new_head = (self.snake_body[0][0], self.map_width - 1)

            elif self.snake_body[0][1] + action[1] >= self.map_width:
                new_head = (self.snake_body[0][0], 0)
                
            else:
                new_head = (self.snake_body[0][0], self.snake_body[0][1] + action[1])
        
        return ((new_head in self.snake_body) and new_head != self.snake_body[snake_length - 1]) \
                or ((new_head in self.snake_body) and new_head == self.snake_body[snake_length - 1] \
                and snake_length == 2)
        
    def evaluate_actions(self):
        actions = []
        snake_length = len(self.snake_body)
        if not self.self_biting((-1, 0)):
            actions.append((-1, 0))
        
        if not self.self_biting((0, +1)):
            actions.append((0, +1))
        
        if not self.self_biting((+1, 0)):
            actions.append((+1, 0))
        
        if not self.self_biting((0, -1)):
            actions.append((0, -1))
        return actions
    
    def get_next_state(self, action):
        snake_body = copy.deepcopy(self.snake_body)
        seeds = copy.deepcopy(self.seeds)

        if action[1] == 0:
            if snake_body[0][0] + action[0] < 0:
                new_head = (self.map_height - 1, self.snake_body[0][1])
                    
            elif snake_body[0][0] + action[0] >= self.map_height:
                new_head = (0, self.snake_body[0][1])
                            
            else:
                new_head = (self.snake_body[0][0] + action[0], self.snake_body[0][1])
                
        else:
            if snake_body[0][1] + action[1] < 0:
                new_head = (self.snake_body[0][0], self.map_width - 1)

            elif snake_body[0][1] + action[1] >= self.map_width:
                new_head = (self.snake_body[0][0], 0)

            else:
                new_head = (self.snake_body[0][0], self.snake_body[0][1] + action[1])
        
        snake_body.insert(0, new_head)
        if (new_head[0], new_head[1], 1) in seeds:
            seeds.remove((new_head[0], new_head[1], 1))
        elif (new_head[0], new_head[1], 2) in seeds:
            seeds.remove((new_head[0], new_head[1], 2))
            seeds.add((new_head[0], new_head[1], 1))
        else:
            snake_body.pop()
        
        return State(snake_body, seeds, self.map_width, self.map_height)

    
class Node:
    def __init__(self, state, parent, path_cost):
        self.state = state
        self.parent = parent
        self.last_action = None
        self.path_cost = path_cost

    def set_last_action(self, last_action):
        self.last_action = last_action

def print_frontier(frontier):
    for node in frontier:
        print("Snake body: ", node.state.snake_body)
        print("Seeds: ", len(node.state.seeds))
        print("Actions: ", node.state.evaluate_actions())
        print("Path depth: ", node.path_cost)
        print("___________")
        
def print_path(node):
    
    parent = node
    path = ""
    while parent:
        if parent.last_action == (-1, 0):
            path += "U-"
        elif parent.last_action == (0, 1):
            path += "R-"
        elif parent.last_action == (1, 0):
            path += "D-"
        elif parent.last_action == (0, -1):
            path += "L-"
        parent = parent.parent
    
    print(path[:-1][::-1])
        
def LDFS(problem, depth):
    root = Node(problem.initial_state, None, 0)
    
    if problem.goal_test(root.state):
        print("Path length: ", root.path_cost)
        print("Path: ", end="")
        print_path(root)
        return True
    
    frontier = [root]
    explored = set()
    
    while frontier:
        node = frontier.pop()
        
        if (str(node.state.snake_body), str(node.path_cost)) not in explored:
            explored.add((str(node.state.snake_body), str(node.path_cost)))

            if node.path_cost >= depth:
                continue
            
            for action in node.state.evaluate_actions():
                child = Node(node.state.get_next_state(action), node, node.path_cost + 1)
                child.set_last_action(action)
                if problem.goal_test(child.state):
                    print("Path length: ", child.path_cost)
                    print("Path: ", end="")
                    print_path(child)
                    return True
                frontier.append(child)
    return False

                    
def IDS(problem):
    depth = 1
    while True:
        result = LDFS(problem, depth)
        if result:
            return result
        depth += 1
    
    return False
    
                
problem = Problem()
problem.get_input2("test3.txt")

time1 = time.time()
IDS(problem)
time2 = time.time()

print("Time: ", (time2 - time1)*1000, " ms")
```

```python
import copy
import time
import bisect

class Problem:
    def get_input2(self, file_path):
        file = open(file_path)
        map_height, map_width = map(int,file.readline().split(','))
        
        snake_head = tuple(map(int,file.readline().split(',')))
        snake_body = [snake_head]
        
        seed_number = int(file.readline())
        
        seeds = set()
        for i in range(seed_number):
            seeds.add(tuple(map(int,file.readline().split(','))))
        
        self.initial_state = State(snake_body, seeds, map_width, map_height)
    
    def get_input1(self):
        map_height, map_width = map(int,input().split(','))
        
        snake_head = tuple(map(int,input().split(',')))
        snake_body = [snake_head]
        
        seed_number = int(input())
        
        seeds = set()
        for i in range(seed_number):
            seeds.add(tuple(map(int,input().split(','))))
        
        self.initial_state = State(snake_body, seeds, map_width, map_height)
    
    def goal_test(self, state):
        return len(state.seeds) == 0

    
class State:
    def __init__(self, snake_body, seeds, map_width, map_height):
        self.snake_body = snake_body
        self.seeds = seeds
        self.map_width = map_width
        self.map_height = map_height
    
    def __eq__(self, other):
        return self.snake_body == other.snake_body and self.seeds == other.seeds
    
    def __hash__(self):
        return hash((str(self.snake_body), str(self.seeds)))
    
    def self_biting(self, action):
        snake_length = len(self.snake_body)
        
        if action[1] == 0:
            if self.snake_body[0][0] + action[0] < 0:
                new_head = (self.map_height - 1, self.snake_body[0][1])

            elif self.snake_body[0][0] + action[0] >= self.map_height:
                new_head = (0, self.snake_body[0][1])

            else:
                new_head = (self.snake_body[0][0] + action[0], self.snake_body[0][1])

        else:
            if self.snake_body[0][1] + action[1] < 0:
                new_head = (self.snake_body[0][0], self.map_width - 1)

            elif self.snake_body[0][1] + action[1] >= self.map_width:
                new_head = (self.snake_body[0][0], 0)
                
            else:
                new_head = (self.snake_body[0][0], self.snake_body[0][1] + action[1])
        
        return ((new_head in self.snake_body) and new_head != self.snake_body[snake_length - 1]) \
                or ((new_head in self.snake_body) and new_head == self.snake_body[snake_length - 1] \
                and snake_length == 2)
        
    def evaluate_actions(self):
        actions = []
        snake_length = len(self.snake_body)
        if not self.self_biting((-1, 0)):
            actions.append((-1, 0))
        
        if not self.self_biting((0, +1)):
            actions.append((0, +1))
        
        if not self.self_biting((+1, 0)):
            actions.append((+1, 0))
        
        if not self.self_biting((0, -1)):
            actions.append((0, -1))
        return actions
    
    def get_next_state(self, action):
        snake_body = copy.deepcopy(self.snake_body)
        seeds = copy.deepcopy(self.seeds)

        if action[1] == 0:
            if snake_body[0][0] + action[0] < 0:
                new_head = (self.map_height - 1, self.snake_body[0][1])
                    
            elif snake_body[0][0] + action[0] >= self.map_height:
                new_head = (0, self.snake_body[0][1])
                            
            else:
                new_head = (self.snake_body[0][0] + action[0], self.snake_body[0][1])
                
        else:
            if snake_body[0][1] + action[1] < 0:
                new_head = (self.snake_body[0][0], self.map_width - 1)

            elif snake_body[0][1] + action[1] >= self.map_width:
                new_head = (self.snake_body[0][0], 0)

            else:
                new_head = (self.snake_body[0][0], self.snake_body[0][1] + action[1])
        
        snake_body.insert(0, new_head)
        if (new_head[0], new_head[1], 1) in seeds:
            seeds.remove((new_head[0], new_head[1], 1))
        elif (new_head[0], new_head[1], 2) in seeds:
            seeds.remove((new_head[0], new_head[1], 2))
            seeds.add((new_head[0], new_head[1], 1))
        else:
            snake_body.pop()
        
        return State(snake_body, seeds, self.map_width, self.map_height)

    
class Node:
    def __init__(self, state, parent, path_cost):
        self.state = state
        self.parent = parent
        self.last_action = None
        self.path_cost = path_cost

    def __lt__(self, other):
        self_seed = max(self.state.seeds, key=lambda x: min(abs(x[0]-self.state.snake_body[0][0]) + abs(x[1] - self.state.snake_body[0][1]) \
                    , abs(self.state.map_height - (x[0] - self.state.snake_body[0][0])) + abs(x[1] - self.state.snake_body[0][1]) \
                    , abs(x[0] - self.state.snake_body[0][0]) + abs(self.state.map_width - (x[1] - self.state.snake_body[0][1])) \
                    , abs(self.state.map_height - (x[0] - self.state.snake_body[0][0])) + abs(self.state.map_width - (x[1] - self.state.snake_body[0][1]))))
            
        self_way = min(abs(self_seed[0]-self.state.snake_body[0][0]) + abs(self_seed[1] - self.state.snake_body[0][1]) \
                    , abs(self.state.map_height - (self_seed[0] - self.state.snake_body[0][0])) + abs(self_seed[1] - self.state.snake_body[0][1]) \
                    , abs(self_seed[0] - self.state.snake_body[0][0]) + abs(self.state.map_width - (self_seed[1] - self.state.snake_body[0][1])) \
                    , abs(self.state.map_height - (self_seed[0] - self.state.snake_body[0][0])) + abs(self.state.map_width - (self_seed[1] - self.state.snake_body[0][1])))
        
        other_seed = max(other.state.seeds, key=lambda x: min(abs(x[0] - other.state.snake_body[0][0]) + abs(x[1] - other.state.snake_body[0][1]) \
                    , abs(other.state.map_height - (x[0] - other.state.snake_body[0][0])) + abs(x[1] - other.state.snake_body[0][1]) \
                    , abs(x[0] - other.state.snake_body[0][0]) + abs(other.state.map_width - (x[1] - other.state.snake_body[0][1])) \
                    , abs(other.state.map_height - (x[0] - other.state.snake_body[0][0])) + abs(other.state.map_width - (x[1] - other.state.snake_body[0][1]))))
        
        other_way = min(abs(other_seed[0] - other.state.snake_body[0][0]) + abs(other_seed[1] - other.state.snake_body[0][1]) \
                    , abs(other.state.map_height - (other_seed[0] - other.state.snake_body[0][0])) + abs(other_seed[1] - other.state.snake_body[0][1]) \
                    , abs(other_seed[0] - other.state.snake_body[0][0]) + abs(other.state.map_width - (other_seed[1] - other.state.snake_body[0][1])) \
                    , abs(other.state.map_height - (other_seed[0] - other.state.snake_body[0][0])) + abs(other.state.map_width - (other_seed[1] - other.state.snake_body[0][1])))
        
        return self.path_cost + self_way > other.path_cost + other_way
   
    def __gt__(self, other):
        return not self.__lt__(other)

        
    def set_last_action(self, last_action):
        self.last_action = last_action

def print_frontier(frontier):
    print("frontier size", len(frontier))
    for node in frontier:
        print("Snake body: ", node.state.snake_body)
        print("Seeds: ", len(node.state.seeds))
        print("Actions: ", node.state.evaluate_actions())
        print("Path depth: ", node.path_cost)
        print("___________")
        
def print_path(node):
    parent = node
    path = ""
    while parent:
        if parent.last_action == (-1, 0):
            path += "U-"
        elif parent.last_action == (0, 1):
            path += "R-"
        elif parent.last_action == (1, 0):
            path += "D-"
        elif parent.last_action == (0, -1):
            path += "L-"
        parent = parent.parent
    
    print(path[:-1][::-1])

def find_index(frontier, child):
    for i in range(len(frontier) - 1, -1 , -1):
        if child.path_cost + len(child.state.seeds) <= frontier[i].path_cost + len(frontier[i].state.seeds):
            return i + 1
    return 0

def Astar(problem):
    root = Node(problem.initial_state, None, 0)

    if problem.goal_test(root.state):
        print("Path length: ", root.path_cost)
        print("Path: ", end="")
        print_path(root)
        return True
    
    frontier = [root]
    explored = set()

    while frontier:
        node = frontier.pop()
        if str(node.state.snake_body) not in explored:
            explored.add(str(node.state.snake_body))

            for action in node.state.evaluate_actions():
                child = Node(node.state.get_next_state(action), node, node.path_cost + 1)
                if str(child.state.snake_body) in explored:
                    continue
                child.set_last_action(action)
                if problem.goal_test(child.state):
                    print("Path length: ", child.path_cost)
                    print("Path: ", end="")
                    print_path(child)
                    return True
                bisect.insort(frontier, child)
            
    return False
    
                
problem = Problem()
problem.get_input2("test1.txt")

time1 = time.time()
Astar(problem)
time2 = time.time()

print("Time: ", (time2 - time1)*1000, " ms")
```
